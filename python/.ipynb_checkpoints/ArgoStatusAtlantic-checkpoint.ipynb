```python
import sys 
sys.path.append("/Users/pvb/Dropbox/Oceanografia/LibreriasPython")
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import netCDF4 
import ncdump
import datetime as dt
```

```python
data_dir='/Users/pvb/Data/Argo/geo/atlantic_ocean'
fecha_inicio=dt.date(2018,1,30)

fig = plt.figure()

for iday in range(0,30):
    ifecha=fecha_inicio-dt.timedelta(days=iday)
    fichero_diario='{}/{}_prof.nc'.format(data_dir,ifecha.strftime("%Y/%m/%Y%m%d"))
    #print(fichero_diario)
    nc_fid = netCDF4.Dataset(fichero_diario, 'r')  # Dataset is the class behavior to open the file
    lats = nc_fid.variables['PLATFORM_NUMBER'][:]  # extract/copy the data    
    lats = nc_fid.variables['LATITUDE'][:]  # extract/copy the data
    lons = nc_fid.variables['LONGITUDE'][:]
    juld = netCDF4.num2date(nc_fid.variables['JULD'][:],units=nc_fid.variables['JULD'].units)
    pres = nc_fid.variables['PRES'][:]
    platform_number=netCDF4.chartostring(nc_fid.variables['PLATFORM_NUMBER'][:]).astype(np.int32)
    
plt.plot(lons,lats,'.')        
```
